```python
df_health = df_health.drop('Unnamed: 0', axis=1)
df_health
```

```python
info=pd.DataFrame(df_health.dtypes).T.rename(index={0:'Data type'})
info=info.append(pd.DataFrame(df_health.isna().sum()).T.rename(index={0:'Null values count'}))
info=info.append(pd.DataFrame(df_health.isna().sum()/df_health.shape[0]*100).T.
                         rename(index={0:'Null values (%)'}))
info=info.append(pd.DataFrame(df_health.min()).T.rename(index={0:'Min values'}))
info=info.append(pd.DataFrame(df_health.max()).T.rename(index={0:'Max values'}))
info
```

```python
temp = df_health["Wardname"].str.split(" - ", n = 1, expand = True)
 
df_health["District"]= temp[0]
 
# making separate last name column from new data frame
df_health["Ward name"]= temp[1]
 
# Dropping old Name columns
df_health.drop(columns =["Wardname"], inplace = True)
 
df_health
```

Population2011Census in df_health and df_environment are not in the same data type, so the one in df_health will be turnen into int64.

```python
df_health['Population2011Census'] = df_health['Population2011Census'].astype(int)
info=pd.DataFrame(df_health.dtypes).T.rename(index={0:'Data type'})
info=info.append(pd.DataFrame(df_health.isna().sum()).T.rename(index={0:'Null values count'}))
info=info.append(pd.DataFrame(df_health.isna().sum()/df_health.shape[0]*100).T.
                         rename(index={0:'Null values (%)'}))
info=info.append(pd.DataFrame(df_health.min()).T.rename(index={0:'Min values'}))
info=info.append(pd.DataFrame(df_health.max()).T.rename(index={0:'Max values'}))
info
```

There are 622 data in `df_health` and 624 data in `df_environment`. So if both tables are joined on `Popolation2011Census`, there will be 2 null values the new joined table.

```python
envi_health=pd.merge(df_health,df_environment,on='Population2011Census', how='right')
info=pd.DataFrame(envi_health.dtypes).T.rename(index={0:'Data type'})
info=info.append(pd.DataFrame(envi_health.isna().sum()).T.rename(index={0:'Null values count'}))
info
```

```python
envi_health[envi_health['GeneralFertilityRate'].isna()]
```
